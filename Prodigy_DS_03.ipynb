```python
columns=df[['age','duration','campaign']]
Q1=columns.quantile(0.25)
Q3=columns.quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
Upper_Limit=(1.5*IQR)+Q3
Lower_Limit=Q1-(1.5*IQR)
```

```python
df[['age','duration','campaign']]=columns[(columns>Lower_Limit) & (columns<Upper_Limit)]
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_split=15)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
result=pd.DataFrame({'Actual_Value':y_test,'Predicted_Vale':y_pred})
result
```

```python
train_score=model.score(x_train,y_train)
test_score=model.score(x_test,y_test)
print("Training Score: ",train_score)
print("\n\nTesting Score: ",test_score)
```

```python
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
print("Confusion Matrix: \n",confusion_matrix(y_test,y_pred))
```

```python
print("Classification Report: \n",classification_report(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(10,10))
plot_tree(model,feature_names=x_train.columns,class_names=['no', 'yes'],filled=True,rounded=True,fontsize=2)
plt.show()
```

```python
model1=DecisionTreeClassifier(criterion='entropy',max_depth=4,min_samples_split=15)
model1.fit(x_train,y_train)
```

```python
y_pred1=model1.predict(x_test)
```

```python
result=pd.DataFrame({'Actual_Value':y_test,'Predicted_Vale':y_pred1})
result
```

```python
train_score=model1.score(x_train,y_train)
test_score=model1.score(x_test,y_test)
print("Training Score: ",train_score)
print("\n\nTesting Score: ",test_score)
```

```python
print("Accuracy Score: ",accuracy_score(y_test,y_pred1))
```

```python
print("Confusion Matrix: \n",confusion_matrix(y_test,y_pred1))
```
